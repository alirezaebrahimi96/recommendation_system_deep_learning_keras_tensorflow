```python
num_unique_users=len(set(list(df_train.userId.unique()) + list(df_test.userId.unique())))
num_unique_items=len(set(list(df_train.itemId.unique()) + list(df_test.itemId.unique())))
```

```python
EMBEDDING_SIZE = 50
users_input = keras.Input(shape=(1,), name="users_input")
users_embedding = keras.layers.Embedding(num_unique_users + 1, EMBEDDING_SIZE, name="users_embeddings")(users_input)
users_bias = keras.layers.Embedding(num_unique_users + 1, 1, name="users_bias")(users_input)

items_input = keras.Input(shape=(1,), name="movies_input")
items_embedding = keras.layers.Embedding(num_unique_items + 1, EMBEDDING_SIZE, name="movies_embedding")(items_input)
items_bias = keras.layers.Embedding(num_unique_items + 1, 1, name="movies_bias")(items_input)

dot_product_users_items = keras.layers.multiply([users_embedding, items_embedding])
input_terms = dot_product_users_items + users_bias + items_bias
input_terms = keras.layers.Flatten(name="fl_inputs")(input_terms)
output = keras.layers.Dense(1, activation="relu", name="output")(input_terms)
model = keras.Model(inputs=[users_input, items_input], outputs=output)
```

```python
model.summary()
```

```python
opt_adam = tf.keras.optimizers.Adam(lr = 0.005)
model.compile(optimizer=opt_adam, loss= ['mse'], metrics=[tf.keras.metrics.MeanSquaredError()])
```

```python
model.fit(x=[df_train.userId.values, df_train.itemId.values], y=df_train.rating.values, batch_size=512, epochs=3, verbose=1, validation_data=([df_test.userId.values, df_test.itemId.values], df_test.rating.values))
```
